```python
import openpyxl
from itertools import chain
import pandas as pd
import numpy as np
import json
from tqdm.notebook import trange, tqdm
```

```python
df3 = pd.read_excel('./test.xlsx').astype(object)
```

```python
flare = {"name": "flare", "children": []}


# iterate through dataframe values
for row in tqdm(df3.values):
    level0 = row[0]
    level1 = row[1]
    level2 = row[2]
    level3 = row[3]
    level4 = row[4]
    value = row[5]
    
    # create a dictionary with all the row data
    d = {'name': level0,
          'children': [{'name': level1,
                        'children': [{'name': level2,
                                          'children': [{'name': level3,  
                                                'children': [{'name': level4, 'value': value}]}]}]}]}
    # initialize key lists
    key_list = []
    child_list = []
    grandchild_list = []
    grandgrandchild_list = []
    #grandgrandgrandchild_list = []

    # iterate through first level node names
    for item in flare['children']:
        key_list.append(item['name'])
        # iterate through next level node names
        child_list = []
        #item.items - главная проблема, видимо 
        for _, v1 in item.items():
            if isinstance(v1, list):
                for x1 in v1:
                    child_list.append(x1['name'])
                    
                    grandchild_list = []
                
                    for _, v2 in item.items():
                        if isinstance(v2, list):
                            for x2 in v2:
                                grandchild_list.append(x2['name'])
                                
                                grandgrandchild_list = []
                                
                                for _, v3 in item.items():
                                    if isinstance(v3, list):
                                        for x3 in v3:
                                            grandgrandchild_list.append(x3['name'])

    # add the full row of data if the root is not in key0
    if level0 not in key_list:
        d = {'name': level0,
              'children': [{'name': level1,
                        'children': [{'name': level2,
                                          'children': [{'name': level3,  
                                                  'children': [{'name': level4, 'value': value}]}]}]}]}
        flare['children'].append(d)

    
    elif level1 not in child_list:

        # if the root exists, then append only the next level children
        d = {'name': level1,
                        'children': [{'name': level2,
                                          'children': [{'name': level3,  
                                                  'children': [{'name': level4, 'value': value}]}]}]}

        flare['children'][key_list.index(level0)]['children'].append(d)
        
    elif level2 not in grandchild_list:
        # if the root exists, then only append the next level children 
        d = {'name': level2,
                    'children': [{'name': level3,  
                            'children': [{'name': level4, 'value': value}]}]}
        
        flare['children'][key_list.index(level0)]['children'][child_list.index(level1)]['children'].append(d)
    
    elif level3 not in grandgrandchild_list:
        # if the root exists, then only append the next level children 
        d = {'name': level3,  
                            'children': [{'name': level4, 'value': value}]}
        
        flare['children'][key_list.index(level0)]['children'][child_list.index(level1)]['children'][grandchild_list.index(level2)]['children'].append(d)
        
    else:
        # if the root exists, then only append the next level children 
        d = {'name': level4, 'value': value}
        
        flare['children'][key_list.index(level0)]['children'][child_list.index(level1)]['children'][grandchild_list.index(level2)]['children'][grandgrandchild_list.index(level3)]['children'].append(d)

print(flare)
# uncomment next three lines to save as json file
lst = json.dumps(flare, indent=2, ensure_ascii=False)
print(lst)
jsonFile = open("test.json", "w")
jsonFile.write(lst)
jsonFile.close()
```
